```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load Pretrained YOLOv5 and YOLOv8 models
yolo_v1 = YOLO("yolov5s.pt")  # YOLOv5 small
yolo_v2 = YOLO("yolov8n.pt")  # YOLOv8 nano
yolo_v3= YOLO("yolov10n.pt")

# Compute FLOPs and Parameters
def compute_model_stats(model, name):
    print(f"\n{name} Model Summary:")
    model.info()  # Correct method for Ultralytics YOLO

compute_model_stats(yolo_v1, "YOLOv5")
compute_model_stats(yolo_v2, "YOLOv8")
compute_model_stats(yolo_v3, "YOLOv10")
# Load an image
image_path = "traffic.jpeg"  # Ensure this image exists

# Run Inference on YOLOv5 and YOLOv8
def run_inference(model, image_path, name):
    results = model(image_path)  # Pass image path
    results[0].show()  # Show bounding box predictions
    plt.imshow(plt.imread(image_path))
    plt.title(f"Predictions using {name}")
    plt.axis("off")
    plt.show()

results_v1 = run_inference(yolo_v1, image_path, "YOLOv5")
results_v2 = run_inference(yolo_v2, image_path, "YOLOv8")
results_v3 = run_inference(yolo_v3, image_path, "YOLOv10")
```
